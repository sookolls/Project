```python
import requests
import pandas as pd
from datetime import datetime, timedelta
from typing import Any
import matplotlib.pyplot as plt

# Параметры
symbol: str = "BTCUSDT"               
product_type: str = "USDT-FUTURES"    
granularity: str = "1H"   # 1 час
days: int = 365
url: str = "https://api.bitget.com/api/v2/mix/market/candles"

# время
end_time: datetime = datetime.now() # текущее время
start_time: datetime = end_time - timedelta(days=days) # начальное время 
# перевод в секунды
end_ms: int = int(end_time.timestamp() * 1000)
start_ms: int = int(start_time.timestamp() * 1000)

all_candles: list[list[Any]] = []
current_end: int = end_ms
max_needed: int = days * 24 #  свечи максимум

while len(all_candles) < max_needed:
    params: dict[str, Any] = {
        "symbol": symbol,
        "productType": product_type,
        "granularity": granularity,
        "endTime": current_end,
        "limit": 1000
    }
    try:
        response: requests.Response = requests.get(url, params=params) #ответ, полученный от сервера после отправки HTTP-запроса
        raw: dict[str, Any] = response.json()

        if raw.get("code") != "00000":
            print("Ошибка API:", raw)
            break

        data: list[list[Any]] = raw.get("data")
        if not data:
            print("Нет данных:", raw)
            break

        all_candles.extend(data)

        print(f"Загружено {len(all_candles)} свечей...")

        oldest_candle: int = int(data[-1][0]) # берём время самой старой свечи
        if oldest_candle <= start_ms:
            break

        current_end = oldest_candle - 1

    except Exception as e:
        print(f"Ошибка при запросе: {e}")
        break

# Преобразуем в DataFrame
df: pd.DataFrame = pd.DataFrame(all_candles, columns=[
    "timestamp", "open", "high", "low", "close", "volume", "quoteVolume"])

# Конвертация timestamp
df["timestamp"] = pd.to_datetime(df["timestamp"].astype(float) / 1000, unit='s', utc=True)

#сортировка
df = df.sort_values("timestamp").reset_index(drop=True)

# Удаляем дубликаты
df = df.drop_duplicates(subset="timestamp").reset_index(drop=True)

# Сохраняем
df.to_csv("candles_1hd.csv", index=False)

print(f"Готово! Всего свечей: {len(df)}")
print(df.head())
print(df.tail())
```

```python
import matplotlib.pyplot as plt

window = int(input("введите ширину окна: "))
df["SMA"] = df["close"].rolling(window).mean()
df["EMA"] = df["close"].ewm(window, adjust=False).mean()

plt.figure(figsize=(12,6))
plt.plot( df["close"].astype("float64"), label="Close Price", color="blue")
plt.plot( df["SMA"], label=f"SMA {window}", color="red")
plt.plot( df["EMA"], label=f"EMA {window}", color="green")

plt.title(f"Почасовой график {symbol}")
plt.xlabel("Время")
plt.ylabel("Цена (USDT)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
