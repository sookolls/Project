```python
import requests
import pandas as pd
from datetime import datetime, timedelta
from typing import Any

# Параметры
symbol: str = "BTCUSDT"               
product_type: str = "USDT-FUTURES"    
granularity: str = "3600"             # 1 час
days: int = 83
url: str = "https://api.bitget.com/api/v2/mix/market/candles"

# время
end_time: datetime = datetime.now() # текущее время
start_time: datetime = end_time - timedelta(days=days) # начальное время (83 дня назад)
# перевод в секунды
end_ms: int = int(end_time.timestamp() * 1000)
start_ms: int = int(start_time.timestamp() * 1000)

all_candles: list[list[Any]] = []
current_end: int = end_ms
max_needed: int = days * 24  # 1992 свечи максимум

while len(all_candles) < max_needed:
    params: dict[str, Any] = {
        "symbol": symbol,
        "productType": product_type,
        "granularity": granularity,
        "endTime": current_end,
        "limit": 1000
    }

    response: requests.Response = requests.get(url, params=params) #ответ, полученный от сервера после отправки HTTP-запроса
    raw: dict[str, Any] = response.json()

    if raw.get("code") != "00000":
        print("Ошибка API:", raw)
        break

    data: list[list[Any]] = raw.get("data")
    if not data:
        print("Нет данных:", raw)
        break

    all_candles.extend(data)

    print(f"Загружено {len(all_candles)} свечей...")

    oldest_candle: int = int(data[-1][0]) # берём время самой старой свечи
    if oldest_candle <= start_ms:
        break

    current_end = oldest_candle

# Преобразуем в DataFrame
df: pd.DataFrame = pd.DataFrame(all_candles, columns=[
    "timestamp", "open", "high", "low", "close", "volume", "quoteVolume"])

# Конвертация timestamp
df["timestamp"] = pd.to_datetime(df["timestamp"].astype(float) / 1000, unit='s', utc=True)

# Удаляем дубликаты
# df = df.drop_duplicates(subset="timestamp").reset_index(drop=True)

# Сохраняем
df.to_csv("candles_1h.csv", index=False)

print(f"Готово! Всего свечей после удаления дубликатов: {len(df)}")
print(df.head())
print(df.tail())

```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(15,6)) #рамзер графика 
plt.plot(df["timestamp"], df["close"], label="Close Price")
plt.title(f"Цена {symbol} за последние {days} дня")
plt.xlabel("Время") 
plt.ylabel("Цена (USDT)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
